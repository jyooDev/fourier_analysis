```python
Y = np.fft.fft(y)/n 
Y = Y[range(int(n/2))]
```

```python
fig, ax = plt.subplots(1, 1, figsize=(12,8))
ax.plot(freq, abs(Y), 'r', linestyle=' ', marker='^') 
ax.set_xlabel('Freq (Hz)')
ax.set_ylabel('|Y(freq)|')
ax.vlines(freq, [0], abs(Y))
ax.set_xlim([0, 20]); ax[1].grid(True)
plt.show()
```
