```python
import numpy as np
import matplotlib.pyplot as plt
import os
import librosa
import IPython.display as ipd 
```

```python
BASE_DIR = "/Users/gina0/Desktop/fourier_analysis/capstone/sound_folder"
piano_c5_file = "piano_c5.wav"
```

```python
ipd.Audio(os.path.join(BASE_DIR, piano_c5_file))
```

```python
piano_c5_data, sr = librosa.load(os.path.join(BASE_DIR, piano_c5_file)) #sr = 22050 
```

```python
# Plot time domain graph
time = np.arange(0, len(piano_c5_data)) / sr
plt.plot(time, piano_c5_data)
plt.xlabel('Time (s)')
plt.ylabel('Amplitude')
plt.title('Audio Waveform (Piano C5)')
plt.show()
```

```python
# apply FFT
piano_c5_fft = np.fft.fft(piano_c5_data)
ms_piano_c5_fft = np.abs(piano_c5_fft)
phase = np.angle(piano_c5_fft)
```

```python
piano_c5_fft[0] #in complex number since fft is applied
ms_piano_c5_fft[0] #in real number (magnitude spectrum)
phase[8]
```

```python
# Plot freq domain graph
freq = np.linspace(0, sr, len(ms_piano_c5_fft)) #[start, stop] num evenly spaced
plt.figure(figsize=(18,5))
plt.plot(freq[:5000], ms_piano_c5_fft[:5000])
plt.xlabel('Freq (f)')
plt.ylabel('Magnitude')
plt.title('FFT (Piano C5)')
plt.show()
```

```python
threshold = 200
ms_piano_c5_fft_filtered = np.where(ms_piano_c5_fft < threshold, 0, ms_piano_c5_fft)
```

```python
plt.figure(figsize=(18,5))
plt.plot(freq[:5000], ms_piano_c5_fft_filtered[:5000])
plt.xlabel('Freq (f)')
plt.ylabel('Magnitude')
plt.title('FFT (Piano C5)')
plt.show()
```

```python
# Inverst FFT from freq to time domain
piano_c5_data_filtered = np.fft.ifft(ms_piano_c5_fft_filtered)
plt.plot(time, piano_c5_data_filtered)
plt.xlabel('Time (s)')
plt.ylabel('Amplitude')
plt.title('Audio Waveform (Piano C5)')
plt.show()
```
