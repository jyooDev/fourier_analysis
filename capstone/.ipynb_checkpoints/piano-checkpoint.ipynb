## DFT
$$\begin{align}
\Upsilon_p &= \frac{1}{N} \sum_{n=0}^{N-1} y_n e^{-2\pi ipn/N}\\
y_n &= \sum_{p=0}^{N-1}\frac{1}{2}\Upsilon_pe^{2\pi ipn/N}
\end{align}$$

$\Upsilon_p$ is DFT and $y_n$ is IDFT, which corresponds to the discrete signal at $n$ in the time domain. The following code is the application of DFT.
Ultimately, DFT is multiplying the discrete time signal $y_n$ the sines and cosines signal of  different frequencies denoted as variable $p$

```python
# apply DFT with numpy
N = len(piano_c5_clip_1s)
T = N / sr #record length(audio time)
freq = np.linspace(0, sr, N)
dft = np.zeros(N, dtype=complex)

start_time = time.time()
for p in range(0, N, 2):
    for n in range(0, N):
        y_j = piano_c5_data[n]
        w_j = np.exp(-1j*2*np.pi*p*n/N)
        dft[p] += (y_j*w_j)
end_time = time.time()
print(f"Execution time: {end_time - start_time:.4f} seconds")
```

```python
ms_dft = np.abs(dft)
fig, ax = plt.subplots()
plot = plt.plot(freq, ms_dft)
plt.xlabel('Freq (f)')
plt.ylabel('Magnitude')
plt.title('DFT (Piano C5)')
annotation = ax.annotate(
    text='',
    xy=(0,0),
    xytext=(15,15),
    textcoords='offset points',
    bbox={'boxstyle': 'round', 'fc': 'w'},
    arrowprops={'arrowstyle': '->'}
)
annotation.set_visible(False)

def motion_hover(event):
    annotation_visbility = annotation.get_visible()
    if event.inaxes == ax:
        is_contained, annotation_index = scatter.contains(event)
        if is_contained:
            data_point_location = scatter.get_offsets()[annotation_index['ind'][0]]
            annotation.xy = data_point_location

            text_label = '({0:.2f}, {0:.2f})'.format(data_point_location[0], data_point_location[1])
            annotation.set_text(text_label)

            annotation.get_bbox_patch().set_facecolor(cmap(norm(colors[annotation_index['ind'][0]])))
            annotation.set_alpha(0.4)

            annotation.set_visible(True)
            fig.canvas.draw_idle()
        else:
            if annotation_visbility:
                annotation.set_visible(False)
                fig.canvas.draw_idle()
fig.canvas.mpl_connect('motion_notify_event', motion_hover)
plt.show()
```
